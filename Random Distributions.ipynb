## Probability density/mass functions

```python
from scipy import  stats
quantiles = linspace(-0.5,1.5, 100)
plot(quantiles, stats.uniform.pdf(quantiles))
ylim((0, 1.1))
```

```python
quantiles = arange(50)

plot(quantiles, stats.poisson.pmf(quantiles, 0.2), 'o-')
plot(quantiles, stats.poisson.pmf(quantiles, 5), 'o-')
plot(quantiles, stats.poisson.pmf(quantiles, 20), 'o-')
legend(['$\lambda$ = 0.2','$\lambda$ = 5','$\lambda$ = 20'])
```

## Two-dimensional cross-correlation

```python
i = imread('galileo.png')
imshow(i)
```

```python
i.shape
```

```python
i = sum(i.astype(float), axis=2)/3.0
imshow(i)
```

```python
i.shape
```

```python
imshow(i)
colorbar()
```

```python
imshow(i, cmap=cm.gray)
colorbar()
```

```python
i = where(i > 0.9, 0, 1)
imshow(i, cmap='gray')
colorbar()
```

```python
o = imread('o.png')
imshow(o)
colorbar()
```

```python
o = o.astype(float).sum(axis=-1)/3
o = where(o > 0.9, 0, 1)
imshow(o, cmap=cm.gray, interpolation='nearest')
```

```python
from scipy.signal import correlate2d
cc = correlate2d(i, o)
imshow(cc)
colorbar()
gcf().set_figheight(8)
```

```python
imshow(where(cc > 200, 1, 0), interpolation='nearest')
colorbar()
gcf().set_figheight(8)
```

```python
imshow(maximum_filter(cc, (10,10)))
gcf().set_figheight(8)
```

```python
subplot(121)
imshow(maximum_filter(cc, (50,50)))
subplot(122)
imshow(i, cmap=cm.gray)
gcf().set_figheight(8)
```
